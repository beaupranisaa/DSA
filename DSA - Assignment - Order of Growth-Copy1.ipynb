### Answer

```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
import pandas as pd
```

```python
# n is froom 0 to 1000
n = np.arange(1,100001)
# n
```

```python
#y0 = 1
y0 = np.ones(len(n))

#y1(x) = n
y1 = [x for x in n]

#y2(x) = n^2
y2 = [x**2 for x in n]

#y3(x) = sqrt(n)
y3 = [math.sqrt(x) for x in n]

# y4(x)= nlog(n)
y4 = [math.log(x)*x for x in n]

#y5(x) = n^3
y5 = [x**3 for x in n]

#y6(x) = 2^n
y6 = [2**x for x in n]
y6 = list(filter(lambda num: num != 0 and num >0 , y6))
#y6
```

```python
plt.plot(n,y0,label = '1')
plt.plot(n,y1,label = 'n')
plt.plot(n,y2,label = 'n^2')
plt.plot(n,y3,label = 'sqrt(n)')
plt.plot(n,y4,label = 'nlog(n)')
plt.plot(n,y5,label = 'n^3')
plt.plot(n[:62],y6,label = '2^n')
plt.xscale("log")
plt.yscale("log")
plt.ylim(1)
plt.legend()
```
